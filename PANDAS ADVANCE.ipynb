## APPLY with LAMBDA

As the name suggests, applies a function along any axis of the DataFrame. and also can be applied in series.

```python
def checkFunc(x) :
    if x < 10:
        return 'G10'
    elif x < 50:
        return 'G50'
    elif x < 90:
        return 'G90'
    else: 
        return 'No'        
```

```python
data['carheight'].apply(lambda x:checkFunc(x))
```

```python
data['carwidth'].apply(lambda x: checkFunc(x))
```

```python
import datetime
a = datetime.datetime.now()
data['carheight'].apply(checkFunc)
b = datetime.datetime.now()
print(b-a)
```

```python
import datetime
a = datetime.datetime.now()
data['carheight'].map(checkFunc)
b = datetime.datetime.now()
print(b-a)
```

## ApplyMap

```python
s = ['symboling','wheelbase','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg','price']
data[s].applymap(checkFunc).head(100)
```

1.Map is defined in series only
2.map accepts dicts, Series, or functions
3.map is elementwise for Series
4.Apply works in the series or dataframe either column or rows
5.apply accepts only functions
6.Apply with lambda each element over the series or columns

7.applymap is elementwise for DataFrames

```python
def someFunc(x,y):
    try:
        return (x/y)
    except:
        return 0
```

```python
data[['carheight','carwidth']].apply(lambda x: someFunc(*x),axis=1)
```

## Pivot Table

which Sub-Category has Standard Deviation for Sales and Profit one point more than the Mean of Sales and Profit?

```python
savemeanInfo=pd.pivot_table(sd,index = 'Sub-Category',values =['Sales','Profit'], aggfunc ='mean').reset_index()
print(savemeanInfo)
```

```python
saveSTDvarInfo=pd.pivot_table(sd,index = 'Sub-Category',values =['Sales','Profit'], aggfunc ='std').reset_index()
print(saveSTDvarInfo)
```

```python
savemeanInfo.head()
```

```python
saveSTDvarInfo.head()
```

```python
savemeanInfo.columns=['Sub-Category','MeanProfitVar','MeanSalesVar']
saveSTDvarInfo.columns=['Sub-Category','STDProfitVar','STDSalesVar']
```

```python
TableSave= pd.merge(savemeanInfo,saveSTDvarInfo,on ='Sub-Category',how ='left')
```

```python
TableSave['Diff_mean'] = TableSave['MeanProfitVar'] - TableSave['STDProfitVar']
TableSave['Diff_STD'] = TableSave['MeanSalesVar'] - TableSave['STDSalesVar']
```

```python
TableSave
```

```python
TableSave.sort_values(by = ['Diff_STD'] )
```

```python
TableSave.sort_values(by = ['Diff_mean'], ascending = False)
```

## Group By

A groupby operation involves some combination of splitting the object, applying a function, and combining the results. This can be used to group large amounts of data and compute operations on these groups.

```python
sd.groupby('Region').sum()['Profit']
```

```python
sd.groupby(['Region','Customer ID']).sum().head(5)
```

```python
sd.groupby(['Region','Customer ID']).agg({'Sales':'sum','Profit':'sum'}).reset_index()
```

```python
saveToEdit=sd.groupby(['Region','Customer ID']).agg({'Sales':{'samValSales':'sum','varvalSales':'var'},'Profit': 'sum'}).reset_index()
```

```python
saveForExample=sd.groupby(['Region','Customer ID']).agg({'Sales':'sum','Profit':'sum'}).reset_index()
```

```python
saveForExample
```

```python
saveForExample['profitRatio']=saveForExample['Profit']/saveForExample['Sales']
```

```python
saveForExample=saveForExample.sort_values(['Region','profitRatio'],ascending=False)
```

```python
saveForExample=saveForExample.reset_index()
```

```python
saveForExample
```

```python
saveForExample.nlargest(5,'Sales')
```

```python
saveForExample.groupby(['Region']).agg({'profitRatio':{lambda x: x.nlargest(20).mean()}})
```

```python
saveForExample[saveForExample['Region']=='West'].head(20)['profitRatio'].mean()
```
